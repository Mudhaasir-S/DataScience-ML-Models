```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
a=requests.get("https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population")
a.status_code
print(a)
```

```python
a.content
print(a)
```

```python
b=bs(a.content,'html.parser')
print(b)
```

```python
t=b.find('table',attrs={'class':"wikitable sortable sticky-header sort-under mw-datatable col2left col6left"})
print(t)
```

```python
import pandas as pd
d=pd.read_html(str(t))
print(d)
```

```python
e=pd.DataFrame(d[0])
print(e)
```
